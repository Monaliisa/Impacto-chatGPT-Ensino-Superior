```python
import pandas as pd
```

```python
url = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vTspACiPr5T07gm_vzsqZR5jBNz4TqWdrBaphk7KbqEVOvKChhJzZG8kvzbhFC5n1Rwj6s74kUksdXB/pub?output=csv'
```

```python
dados = pd.read_csv(url)
dados
```

```python
dados.info()
```

```python
dados.drop(columns=['Carimbo de data/hora'], inplace=True)
dados
```

```python
dados.columns = ['1.1 Curso ou Pos', '1.2 Faixa etaria', '1.3 Fez uso','1.4 Tipo de uso', '1.5 Assinou a ferramenta?', '2.1 Decisao da adesao', '2.2 Motivacao do uso', '2.3 Frequencia', '3.1 Dificuldades', '3.2 Confianca', '3.3 Utiliza Engenharia de Prompt?', '4.1 Formando em 2023/2024?', '4.2 Utilizou chatGPT ou similiares no TG?', '4.3 Essencial para o desenvolvimento do TG?', '5.1 Impacto chatGPT ou similares estudo', '5.2 Impacto no rendimento academico', '5.3 Aspectos da eficacia', '5.4 Dependencia da ferramenta', '6.1 Experiencia com chatGPT']
```

```python
dados.info()
```

```python
dados.head()
```

```python
dados.fillna('Não respondido', inplace=True)
dados
```

Tratamento coluna '1.4 Tipos de uso'

```python
dados['1.4 Tipo de uso'] = dados['1.4 Tipo de uso'].str.replace(", ", ",")
```

```python
dados_dummies = dados['1.4 Tipo de uso'].str.get_dummies(sep=',')
dados_dummies
```

```python
dados_with_dummies = pd.concat([dados, dados_dummies], axis=1)
dados_with_dummies
```

```python
column_rename_dict = {
    'Uso ou usei ferramentas como o ChatGPT para uso acadêmico': '1.4 Uso_Academico',
    'Uso ou usei ferramentas como o ChatGPT para objetivos pessoais': '1.4 Uso_Pessoal',
    'Uso ou usei ferramentas como o ChatGPT no meu trabalho': '1.4 Uso_Trabalho',
    'Não respondido' : '1.4 Não respondido'
    # Adicione mais renomeações conforme necessário
}
```

```python
dados_with_dummies.rename(columns=column_rename_dict, inplace=True)
dados_with_dummies
```

Tratamento coluna 2.1

```python
quantidade = dados_with_dummies['2.1 Decisao da adesao'].nunique()
quantidade
```

```python
dados_with_dummies['2.1 Decisao da adesao'].value_counts()
```

```python
dados_with_dummies['2.1 Decisao da adesao'] = dados_with_dummies['2.1 Decisao da adesao'].str.replace(", ", ",")
```

```python
colunas_relevantes = [
    'Curiosidade',
    'Recomendação de amigos/colegas',
    'Indicação de professores ou lideranças no trabalho',
    'Necessidade acadêmica/profissional'
]
```

```python
dummies = pd.get_dummies(dados_with_dummies['2.1 Decisao da adesao'])
dummies
```

```python
dummies['Indicação de professores ou lideranças no trabalho'] = False
dummies
```

```python
dummies = dummies[colunas_relevantes]
dummies
```

```python
dados_with_dummies = pd.concat([dados_with_dummies, dummies], axis=1)
dados_with_dummies
```
